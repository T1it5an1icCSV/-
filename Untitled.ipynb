```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('processed_titanic.csv')
```

1. Линейный график; Изменения в цены билетов в зависимости от даты.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Создание диапазона дат
date_range = pd.date_range(start='2023-01-01', end='2023-03-31')

# Генерация случайных цен
np.random.seed(0)  # Для воспроизводимости
random_prices = np.random.uniform(low=100, high=500, size=len(date_range))

# Генерация случайного значения для второго столбца
random_values = np.random.randint(1, 10, size=len(date_range))

# Создание DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Price': random_prices,
    'RandomValue': random_values
})

# Установка столбца с датами в качестве индекса
data.set_index('Date', inplace=True)

# Построение графика
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Price'], label='Цена билета', color='blue')
plt.title('Изменение цены билета по датам')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

***Вывод:***   Из графика видно что цены на Титаник имели довольно большие скачки в их цене.
Например: 
5.01 - 01.02: После пика в середине января цена начинает постепенно снижаться, достигая минимумов около 150 рублей в конце января.
01.02 - 15.02: В начале февраля наблюдается небольшое повышение цен, которое снова колеблется в диапазоне 200-350 рублей

15.02 - 01.03: К концу февраля цена стабилизируется и колеблется в пределах 250-350 рублй.

01.03 - 15.03: В начале марта происходит резкий рост цен, достигая 450 рублей, затем снова наблюдается снижение и колебания между 300-400 рубями.

15.03 - 01.04: В последней части марта цена снова испытывает колебания, но в целом остается выше 300 рублей, с кратковременными подъемами до 450 ожений



Скачки цен на билеты Titanic могут быть объяснены несколькими факторами. Во-первых, сезонность и спрос на путешествия во время определённых прные даты. Во-вторых, ограниченное количество мест на корабле также способствовало росту цен при высокомсажиров.

2. Гистограмма;

```python
# 2. Гистограмма: Распределение возраста пассажиров по классам обслуживания
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Pclass', multiple='stack', kde=False, bins=30)
plt.title('Распределение возраста пассажиров по классам обслуживания')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend(title='Класс обслуживания', labels=['1 класс', '2 класс', '3 класс'])
plt.show()
```

***Вывод:***

15-25 лет: Здесь наблюдается постепенный рост числа пассажиров (100-120 человек) с достаточно равномерным распределением между классами. Это свидетельствует о том, что молодые люди активно путешествовал.

25-35 лет: Наибольшая концентрация пассажиров (около 350 человек) сосредоточена в этом возрастном диапазоне, с пиком около 30 лет. Здесь преобладают пассажиры первого класса, что может указывать на более высокую платежеспособность этой групы.

35-45 лет: Количество пассажиров снижается до 50-70 человек, с более равномерным распределением по классам. Это может говорить о том, что интерес к путешествиям в этом возрасте начинает ослабвать.

50-70 лет: В этой группе количество пассажиров значительно меньше (20-30 человек). Наблюдается легкое преобладание пассажиров первого класса, что может указывать на более высокую платежеспособность пожилы людей.

3. Столбчатый график;
Выживаемость по полу и классу обслуживания

```python
# 3. Столбчатый график: Выживаемость по полу и классу обслуживания
plt.figure(figsize=(10, 6))
survival_by_sex_class = df.groupby(['Pclass', 'Sex'])['2urvived'].mean().unstack()
survival_by_sex_class.plot(kind='bar', stacked=True)
plt.title('Выживаемость по полу и классу обслуживания')
plt.xlabel('Класс обслуживания')
plt.ylabel('Процент выживших')
plt.legend(title='Пол', labels=['Мужчины', 'Женщины'])
plt.show()
```

***Вывод:***  На основе данных из графика, можно сделать вывод, что наиболее высокая выживаемость соотвествует первому классу обслуживанияти - около 90%, при этом доля выживших женщин (около 65%) значительно превышает долю выживших мужчин (около 25%). Это указывает на приоритетное спасение пассажиров первого класса, особенно женщин.

Во втором классе обслуживания общая выживаемость составляет примерно 75%, где также преобладает женская выживаемость (около 65%) над мужской (около 10%). Заметно существенное снижение выживаемости мужчин по сравнению с первым классом.

В третьем классе наблюдается самый низкий показатель общей выживаемости - около 40%, с сохранением тенденции преобладания женской выживаемости (около 30%) над мужской (около 10%). Это свидетельствует о наименее благоприятных условиях спасения для пассажиров третьего кл
Обьянить наиболее высокий процент выживаемости среди пассажиров первого класса в сравнении с другими можно тем, что они имели более выгодный и быстрый доступ к спасательным шлюпкам.гРазницать, что разрыв в выживаемости между мужчинами и женщинами сохраняется во всех классахчетколее драматично проявляется в первом и втором классах, где доля выживших женщин в несколько раз превышает долю выживших мужчин.

4. Круговая диаграмма;
Процент пола и возраста среди выживших

```python
# Подсчет количества выживших по полу
survived_counts = df[df['2urvived'] == 1]['Sex'].value_counts()

# Замена числовых меток на текстовые
labels = ['Мужчины', 'Женщины']  # Метки для полов в правильном порядке

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(survived_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Процент пола среди выживших')
plt.axis('equal')  
plt.show()
```

***Вывод:***  Исходя из данных круговой диаграммы, можно сделать вывод о том, что правило эвакуации, которое действовало на Титанике "Женщины и дети спасаются в первую очередь" действительно существовало и имело место быть во время трагедии. Видно, что 68.1% выживших составляют женщины, в то время как оставшиеся 31.9% состовляют мужчины. Доля мужчин среди выживших значительно ниже, что может указывать указывать на их меньшую защищенность в сравнении с женщинами или их добровольное уступление мест в спасательных средствах.

Зависимость выживаемости от возраста и пола

```python
# 5. Дисперсионный график: Зависимость выживаемости от возраста и пола
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='2urvived', hue='Sex', alpha=0.6)
plt.title('Зависимость выживаемости от возраста и пола')
plt.xlabel('Возраст')
plt.ylabel('Выживаемость (0 = погиб, 1 = выжил)')
plt.grid()
plt.show()
```

***Вывод:*** Из данных графика видно, что наиболее высоким процентом выживания обладают младенцы и маленькие дети в возрасте от 0 до 10 лет. Также, можно заметить, что для данной возрастной группы наличие определенного пола почти никак не сказалось на процент выживаемости, у женщин небольшой перевес в выживаемости.
В возрастной группе молодых людей от 10 до 30 лет наблюдается постепенное снижение выживаемости. Женщины имеют заметно более высокий шанс выжить, чем мужчины.
В возрастной группе от 30 до 50 лет разница между выживаемостью женщин и мужчин становится особенно заметной. Женщины значительно чаще выживают, что, вероятно, связано с социальными нормами того времени. Среди мужчин этой возрастной группы выживаемость минимальна.
В группе пожилых людей (50–80 лет) выживаемость крайне низкая, исключения составляют несколько случаев в возрастной группе около 60 лет, где женщины выживают чаще.Среди мужчин этой возрастной группы выживаемость гараздо меньше.
Для возрастов 70–80 лет почти все пассажиры погибли, за исключением единичных случаев с женщинами.
